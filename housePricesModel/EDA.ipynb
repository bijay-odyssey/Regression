## Missing Values Analysis

### Checking Values

```python
missing =  house.isnull().sum().sort_values(ascending=False)
print(missing[missing>0])
```

### Visualizing Missing Values

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.heatmap(house.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

## Distribution Visualization

```python
# Histogram and KDE for SalePrice
plt.figure(figsize=(10, 5))
sns.histplot(house['SalePrice'], kde=True)
plt.title('SalePrice Distribution')
plt.show()
```

## Categorical Variable Exploration

```python
# Bar plot for Neighborhood
plt.figure(figsize=(12, 5))
sns.countplot(x='Neighborhood', data=house)
plt.xticks(rotation=45)
plt.title('Neighborhood Distribution')
plt.show()
```

## Outlier Detection

```python
import numpy as np
# Log-scale scatterplot for SalePrice vs GrLivArea
plt.figure(figsize=(12, 6))
plt.scatter(house['GrLivArea'], house['SalePrice'].apply(np.log))
plt.xlabel('GrLivArea')
plt.ylabel('Log(SalePrice)')
plt.title('Log-scale SalePrice vs GrLivArea')
plt.show()
```

```python
# IQR for SalePrice
Q1 = house['SalePrice'].quantile(0.25)
Q3 = house['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
outliers = house[(house['SalePrice'] < (Q1-1.5*IQR)) | (house['SalePrice']> (Q3+1.5*IQR))]
print(f"Number of SalePrice outliers: {len(outliers)}")
```

## Correlation Analysis

```python
# Check top correlated features with SalePrice
correlations = house.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
print(correlations.head(10))
```

```python
# Correlation heatmap for numerical features
corr = house.select_dtypes(include=['int64', 'float64']).corr()
plt.figure(figsize=(50, 40))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Relationship Exploration

```python
# Violin plot for SalePrice by OverallQual
sns.violinplot(x='OverallQual', y='SalePrice', data=house)
plt.title('SalePrice by OverallQual')
plt.show()
```
